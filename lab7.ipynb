# My_Zip()

```python
def my_zip(struct):
    names=("Aagman","Deepamshu","Shubham","Amrit")
    cust_id=(68,69,65,17)
    points=(100,250,145,225)
    zipped=[]
    if(struct):
        if(len(names)==len(cust_id) and len(names)==len(points)):
            for i in range (len(names)):
                zipped.append((names[i],cust_id[i],points[i]))
        else:
            return None
    else:
        for i in range (min(len(cust_id),len(names),len(points))):
            zipped.append((names[i],cust_id[i],points[i]))
    return zipped
print(my_zip(True))
            




```

# my_sort() function


```python
import copy
def mysort(zipped,x):
    sorted=copy.deepcopy(zipped)
    for i in range(len(zipped)):
        for j in range(i+1,len(zipped)):
            if(zipped[i][x]<zipped[j][x]):
                zipped[i],zipped[j]=zipped[j],zipped[i]
    return sorted
print(mysort(my_zip(True),1))
```

# mymax() function

```python
def mymax(a):
    maxi=float("-inf")
    for i in a:
        if i>maxi:
            maxi=i
    return maxi
print(mymax([-4,1,2,3,-7,0,1]))
print(mymax((1,2,5,6,7,8,91,111,25,15)))
print(mymax({9,8,2,1,5,7}))
```

```python
import random
import numpy as np
import matplotlib.pyplot as plt

def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

def generate_numbers(k, n):
    random_numbers = random.sample(range(1, n + 1), k)
    primes = [num for num in random_numbers if is_prime(num)]
    composites = [num for num in random_numbers if not is_prime(num) and num > 1]
    return primes, composites

K = int(input("Enter the number of random numbers (minimum 10): "))
while K < 10:
    K = int(input("K must be at least 10. Please enter again: "))
N = int(input("Enter the upper limit N: "))

primes, composites = generate_numbers(K, N)

prime_squares = [p**2 for p in primes]
composite_roots = [np.sqrt(c) for c in composites]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

ax1.scatter(primes, prime_squares, color='blue', label='Prime Numbers')
ax1.set_title('Prime Numbers vs Their Squares')
ax1.set_xlabel('Prime Numbers')
ax1.set_ylabel('Squares of Prime Numbers')
ax1.grid()
ax1.legend()

ax2.scatter(composites, composite_roots, color='red', label='Composite Numbers')
ax2.set_title('Composite Numbers vs Their Square Roots')
ax2.set_xlabel('Composite Numbers')
ax2.set_ylabel('Square Roots of Composite Numbers')
ax2.grid()
ax2.legend()

plt.tight_layout()
plt.show()
```
